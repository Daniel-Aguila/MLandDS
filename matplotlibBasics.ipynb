```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y=2*x
```

to plot we do plot(x,y)

In Jupyter Notebook they all have to go in the same cell

<h1>Basics</h1>

```python
plt.plot(x,y)
plt.xlabel('X Axis') #label for X axis
plt.ylabel('Y Axis') #label for Y axis
plt.xlim(0,6) #set a x limit between 0,6
plt.ylim(0,15) #set a y limit between 0,15
plt.title('String Title'); #we add a ; so we don't see the return object that we normally see
plt.savefig('saveplot.png') #to save figure
```

<h1>Figure Object</h1>

plt.figure() creates a figure empty size 432x288

```python
plt.figure(figsize=(10,10)) #to set the figure size
```

we need axes to plot on and we can add as many as possible

Using axes, we can actually have multiple plots on top of each other, although normally we use subplots

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) #add_axes([left_bottom,right_bottom,width,height]) #to set the figure size/layout
axes.plot(x,y);
```

```python
#Data 
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig = plt.figure()

#Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small Axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

<h3>Figure Parameters and Exporting a Figure</h3>

```python
fig = plt.figure(figsize=(10,3),dpi=300) #if the image seems blurry we can increase the dpi
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('figure.png',bbox_inches='tight') #bbox_inches fits it correctly including the x and y labels and axis
```

Scatter plots show the relationships between two continuous features, not categorical.

```python
import seaborn as sns
```

```python
df = pd.read_csv('DATA/dm_office_sales.csv')
```

<h3>Hue</h3>

choosing categorical data for hue

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',size='work experience')
#hue colors the points by different values base on the category
#we can base the size on another column
```

choosing continuous data for hue

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary',s=100)
#if one chooses a continuous data, it creates a gradient between the values
#s= it's just a generative size
```

alpha sets transparency

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary',s=100,alpha=0.2)
```

<h3>Style</h3>

it changes the style of the plots by a categorical feature

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)
#if one chooses a continuous data, it creates a gradient between the values
#s= it's just a generative size
```

<h1>Distribution Plots</h1>

3 main distribution plot types

<ul>
    <li>Rug Plot - adds a dash or a tick line for every single value, y-axis doesn't really have meaning</li>
    <li>Histogram - builds of Rug Plot and it is more readible, and y-axis can be normalized as a percentage or as a count</li>
    <li>KDE Plot - Kernel Density Estimation, it is a method of estimating a probability density function of a random variable</li>
</ul>

```python
sns.rugplot(x='salary',data=df,height=0.5)#height does not add value
```

```python
sns.set(style='darkgrid') #you can do whitegrid for a white grid background,other options are ticks,dark,white
#sns.set(style) sets the style for everything in the notebook
sns.displot(data=df,x='salary',bins=10,color='red',edgecolor='blue',lw=4,ls='--',kde=True,rug=True)
```

To only get a kde

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30,kde=True)
```

you can clip it too

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=1,shade=True) #bandwidth with higher value, it picks up less noise or variance
#shade fills in the inside
```

<h1>Categorical Plots</h1>

Display a statistical metric per a category

<ul>
    <li>countplot() - Counts number of rows per category</li>
    <li>barplot() - General for of displaying any chosen metric per category</li>
</ul>

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division',palette='Set2') #base on level of education
#what sort of division are they ideally working in
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')
#calculate the mean salary per level of education #by default estimator is mean
#ci='sd' lets you see standard diviation

plt.legend(bbox_to_anchor=(1.05,1))
```

<h1>Categorical Plots - Distribution within Categories</h1>
<ul>
    <li>Boxplot - displays the distribution of a continuous variable</li>
    <li>Violinplot - displays the probability density across the data using a KDE</li>
    <li>Swarmplot - shows all the data points in the distribution</li>
    <li>Boxenplot(Letter-Value Plot) - designed as an expansion upon the normal box plot</li>
